```python
#lstm  2d

# --- LSTM for PaDEL Descriptors ---
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# --- Step 1: Load data ---
padel_file = r"E:\neuro lab\padel_gui_output_2.csv"
full_df = pd.read_csv(padel_file)

# Merge with your labels
labels = pd.read_excel("abs_Caco2_Wang.xlsx")["Y"]
full_df["Y"] = labels

print(" Data loaded successfully.")
print(full_df.shape)
print(full_df.head())

# --- Step 2: Clean and separate features/target ---
X = full_df.drop(columns=["Y"])
y = full_df["Y"]

# Drop non-numeric or identifier columns (like "Name")
X = X.select_dtypes(include=[np.number])

# Handle NaN
X = X.fillna(X.mean())

# --- Step 3: Scale features ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- Step 4: Reshape for LSTM ---  Reshape the 2D descriptor matrix into a 3D tensor expected by an LSTM:
# (samples, timesteps, features) where in the script we used timesteps = number_of_descriptors and features = 1.
X_scaled = X_scaled.reshape((X_scaled.shape[0], X_scaled.shape[1], 1))

# --- Step 5: Split data ---
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42     # do cross validation 
)

# --- Step 6: Build LSTM model ---   # do hyper parameter tuning
model = Sequential([
    LSTM(64, return_sequences=False, input_shape=(X_train.shape[1], 1)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='linear')   # Use 'sigmoid' if classification (0/1)
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# --- Step 7: Train ---
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)

# --- Step 8: Evaluate ---
loss, mae = model.evaluate(X_test, y_test, verbose=1)
print(f"\n Test MAE: {mae:.4f}")

# --- Step 9: Predict ---
y_pred = model.predict(X_test)
print("\nSample predictions:")
for i in range(5):
    print(f"True: {y_test.iloc[i]:.3f} | Predicted: {y_pred[i][0]:.3f}")   # Xai find features learned by the model
```

```python
# --- LSTM on MACCS Keys (RDKit) ---

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from rdkit import Chem
from rdkit.Chem import MACCSkeys
from tqdm import tqdm

# --- Step 1: Load original data ---
excel_file = r"E:\neuro lab\abs_Caco2_Wang.xlsx"   # make sure path is correct
df = pd.read_excel(excel_file)

# --- Step 2: Use correct column names ---
smiles_col = "Drug"   # this column contains SMILES
y = df["Y"]

print(f" Using SMILES column: {smiles_col}")
print(f" Total molecules: {len(df)}")

# --- Step 3: Generate MACCS keys ---
maccs_data = []

for smi in tqdm(df[smiles_col], desc="Generating MACCS keys"):
    mol = Chem.MolFromSmiles(smi)
    if mol is not None:
        fp = MACCSkeys.GenMACCSKeys(mol)
        maccs_data.append(list(fp))
    else:
        # Handle invalid SMILES (fill with zeros)
        maccs_data.append([0]*167)

# Convert to DataFrame
maccs_df = pd.DataFrame(maccs_data)
maccs_df.columns = [f"MACCS_{i}" for i in range(maccs_df.shape[1])]

# --- Step 4: Merge with Y ---
full_df = maccs_df.copy()
full_df["Y"] = y

print(" MACCS keys generated and merged successfully.")
print(full_df.shape)
print(full_df.head())

# --- Step 5: Prepare features and target ---
X = full_df.drop(columns=["Y"])
y = full_df["Y"]

# --- Step 6: Scale features ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- Step 7: Reshape for LSTM ---
X_scaled = X_scaled.reshape((X_scaled.shape[0], X_scaled.shape[1], 1))

# --- Step 8: Train/test split ---
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# --- Step 9: Build LSTM model ---
model = Sequential([
    LSTM(64, return_sequences=False, input_shape=(X_train.shape[1], 1)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='linear')  # use 'sigmoid' if classification (0/1)
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# --- Step 10: Train ---
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)

# --- Step 11: Evaluate ---
loss, mae = model.evaluate(X_test, y_test, verbose=1)
print(f"\n Test MAE: {mae:.4f}")

# --- Step 12: Predict ---
y_pred = model.predict(X_test)
print("\nSample predictions:")
for i in range(5):
    print(f"True: {y_test.iloc[i]:.3f} | Predicted: {y_pred[i][0]:.3f}")
```

```python
#maccs, 3d, 2d 
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from scipy.stats import spearmanr
from rdkit import Chem
from rdkit.Chem import MACCSkeys
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")

# -----------------------------------------
# PART 1  Load Labels
# -----------------------------------------
label_file = r"E:\neuro lab\abs_Caco2_Wang.xlsx"
df_y = pd.read_excel(label_file)
y = df_y["Y"]
smiles_list = df_y["Drug"]

print("Total molecules:", len(y))


# -----------------------------------------
# PART 2  Generate MACCS Keys
# -----------------------------------------
maccs_data = []

print("\nGenerating MACCS keys")
for smi in tqdm(smiles_list):
    mol = Chem.MolFromSmiles(smi)
    if mol:
        fp = MACCSkeys.GenMACCSKeys(mol)
        maccs_data.append(list(fp))
    else:
        maccs_data.append([0]*167)

maccs_df = pd.DataFrame(maccs_data)
maccs_df.columns = [f"MACCS_{i}" for i in range(maccs_df.shape[1])]
print("MACCS Shape:", maccs_df.shape)


# -----------------------------------------
# PART 3  Load 2D PaDEL descriptors
# -----------------------------------------
padel_2d_file = r"E:\neuro lab\padel_gui_output_2.csv"
padel2d_df = pd.read_csv(padel_2d_file)
padel2d_df = padel2d_df.select_dtypes(include=[np.number]).fillna(0)
padel2d_df.columns = [f"PAD2D_{c}" for c in padel2d_df.columns]
print("2D PaDEL Shape:", padel2d_df.shape)


# -----------------------------------------
# PART 4  Load 3D PaDEL descriptors
# -----------------------------------------
padel_3d_file = r"E:\neuro lab\padel_3d_gui_output.csv"   # correct actual path
padel3d_df = pd.read_csv(padel_3d_file)
padel3d_df = padel3d_df.select_dtypes(include=[np.number]).fillna(0)
padel3d_df.columns = [f"PAD3D_{c}" for c in padel3d_df.columns]
print("3D PaDEL Shape:", padel3d_df.shape)


# -----------------------------------------
# PART 5  MERGE ALL FEATURES
# -----------------------------------------
full_df = pd.concat([maccs_df, padel2d_df, padel3d_df], axis=1)
full_df["Y"] = y

print("\nFinal combined feature matrix:", full_df.shape)


# -----------------------------------------
# PART 6  Train-test preparation
# -----------------------------------------
X = full_df.drop(columns=["Y"])
y = full_df["Y"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reshape for LSTM
X_scaled = X_scaled.reshape((X_scaled.shape[0], X_scaled.shape[1], 1))

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)


# -----------------------------------------
# PART 7  LSTM MODEL
# -----------------------------------------
model = Sequential([
    LSTM(128, return_sequences=False, input_shape=(X_train.shape[1], 1)),
    Dropout(0.35),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='linear')
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=12,
    restore_best_weights=True
)

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=150,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)


# -----------------------------------------
# PART 8  Evaluation
# -----------------------------------------
loss, mae = model.evaluate(X_test, y_test, verbose=1)

y_pred = model.predict(X_test).reshape(-1)

rmse = np.sqrt(np.mean((y_test - y_pred)**2))
spearman_corr, _ = spearmanr(y_test, y_pred)

print("\n============================")
print("      FINAL RESULTS")
print("============================")
print(f"MAE       : {mae:.4f}")
print(f"RMSE      : {rmse:.4f}")
print(f"Spearman  : {spearman_corr:.4f}")

print("\nSample predictions:")
for i in range(5):
    print(f"True: {y_test.iloc[i]:.3f} | Pred: {y_pred[i]:.3f}")
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import MACCSkeys
from mordred import Calculator, descriptors

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error

import torch
import torch.nn as nn
import torch.optim as optim

# ---------------------------------------------------------
# CONFIG
# ---------------------------------------------------------
# PADEL_2D_FILE = "padel_2d.csv"
# PADEL_3D_FILE = "padel_3d.csv"
MORDRED_CACHE = "mordred_features.csv"

MODEL_TYPE = "cnn"     # <-- change to: "cnn", "rnn", "lstm"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Using:", DEVICE)

# ---------------------------------------------------------
# 1. Load label data
# ---------------------------------------------------------
labels_df = pd.read_excel("abs_Caco2_Wang.xlsx")
labels_df["Drug_ID"] = labels_df["Drug_ID"].astype(str)

# ---------------------------------------------------------
# 2. Load PaDEL descriptor files
# ---------------------------------------------------------
padel2d_df = pd.read_csv(PADEL_2D_FILE)
padel3d_df = pd.read_csv(PADEL_3D_FILE)

padel2d_df["Name"] = padel2d_df["Name"].astype(str)
padel3d_df["Name"] = padel3d_df["Name"].astype(str)

# ---------------------------------------------------------
# 3. Merge
# ---------------------------------------------------------
merged = labels_df.merge(padel2d_df, left_on="Drug_ID", right_on="Name", how="inner")
merged = merged.merge(padel3d_df, on="Name", how="inner", suffixes=("_2D", "_3D"))

smiles = merged["Drug"].astype(str)
y = merged["Y"]

# ---------------------------------------------------------
# 4. Convert SMILES
# ---------------------------------------------------------
mols = smiles.apply(Chem.MolFromSmiles)
valid = mols.notna()

merged = merged[valid].reset_index(drop=True)
mols = mols[valid]
y = y[valid].reset_index(drop=True)

# ---------------------------------------------------------
# 5. MACCS
# ---------------------------------------------------------
maccs_df = mols.apply(lambda mol: list(MACCSkeys.GenMACCSKeys(mol))).apply(pd.Series)
maccs_df.columns = [f"MACCS_{i}" for i in range(maccs_df.shape[1])]

# ---------------------------------------------------------
# 6. Mordred descriptors (cache)
# ---------------------------------------------------------
try:
    print("Loading cached Mordred...")
    mordred_df = pd.read_csv(MORDRED_CACHE)
except:
    print("Computing Mordred...")
    calc = Calculator(descriptors, ignore_3D=True)
    mordred_df = calc.pandas(mols)
    mordred_df = mordred_df.apply(pd.to_numeric, errors="coerce").fillna(0)
    mordred_df.to_csv(MORDRED_CACHE, index=False)

# ---------------------------------------------------------
# 7. PaDEL
# ---------------------------------------------------------
padel2d_features = merged.filter(regex="_2D$")
padel3d_features = merged.filter(regex="_3D$")

# ---------------------------------------------------------
# 8. MODELS (CNN, RNN, LSTM)
# ---------------------------------------------------------

# ---------- 3 Layer RNN ----------
class QSAR_RNN(nn.Module):
    def __init__(self, input_size):
        super().__init__()
        self.rnn1 = nn.RNN(input_size, 256, batch_first=True)
        self.rnn2 = nn.RNN(256, 128, batch_first=True)
        self.rnn3 = nn.RNN(128, 64, batch_first=True)
        self.fc = nn.Linear(64, 1)

    def forward(self, x):
        out, _ = self.rnn1(x)
        out, _ = self.rnn2(out)
        out, _ = self.rnn3(out)
        return self.fc(out[:, -1, :])


# ---------- 3 Layer LSTM ----------
class QSAR_LSTM(nn.Module):
    def __init__(self, input_size):
        super().__init__()
        self.lstm1 = nn.LSTM(input_size, 256, batch_first=True)
        self.lstm2 = nn.LSTM(256, 128, batch_first=True)
        self.lstm3 = nn.LSTM(128, 64, batch_first=True)
        self.fc = nn.Linear(64, 1)

    def forward(self, x):
        out, _ = self.lstm1(x)
        out, _ = self.lstm2(out)
        out, _ = self.lstm3(out)
        return self.fc(out[:, -1, :])


# ---------- 3 Layer 1D CNN ----------
class QSAR_CNN(nn.Module):
    def __init__(self, input_size):
        super().__init__()

        # Input shape: (batch, seq=1, features)
        self.conv1 = nn.Conv1d(1, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv1d(32, 64, kernel_size=3, padding=1)
        self.conv3 = nn.Conv1d(64, 128, kernel_size=3, padding=1)

        self.relu = nn.ReLU()
        self.fc = nn.Linear(128 * input_size, 1)

    def forward(self, x):
        x = x.permute(0, 2, 1)  # (batch, features, seq)
        # x: (batch, 1, num_features)

        x = self.relu(self.conv1(x))
        x = self.relu(self.conv2(x))
        x = self.relu(self.conv3(x))
        x = x.flatten(1)
        return self.fc(x)


# ---------------------------------------------------------
# 9. TRAINING FUNCTION
# ---------------------------------------------------------
def evaluate_model(X, y):

    X = X.apply(pd.to_numeric, errors="coerce").fillna(0)

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # shape for RNN/CNN/LSTM
    X_scaled = X_scaled.reshape(-1, 1, X_scaled.shape[1])

    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.2, random_state=42)

    X_train = torch.tensor(X_train, dtype=torch.float32).to(DEVICE)
    X_test = torch.tensor(X_test, dtype=torch.float32).to(DEVICE)
    y_train = torch.tensor(y_train.values, dtype=torch.float32).unsqueeze(1).to(DEVICE)
    y_test_t = torch.tensor(y_test.values, dtype=torch.float32).unsqueeze(1).to(DEVICE)

    input_size = X_train.shape[2]

    # ---------------- SELECT MODEL HERE ----------------
    if MODEL_TYPE == "cnn":
        model = QSAR_CNN(input_size)
    elif MODEL_TYPE == "rnn":
        model = QSAR_RNN(input_size)
    else:
        model = QSAR_LSTM(input_size)
    # ---------------------------------------------------

    model = model.to(DEVICE)

    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    # ----- Train -----
    for epoch in range(35):
        model.train()
        optimizer.zero_grad()

        pred = model(X_train)
        loss = criterion(pred, y_train)
        loss.backward()
        optimizer.step()

    # ----- Predict -----
    model.eval()
    with torch.no_grad():
        preds = model(X_test).cpu().numpy()

    r2 = r2_score(y_test, preds)
    rmse = np.sqrt(mean_squared_error(y_test, preds))

    return {"R2": float(r2), "RMSE": float(rmse)}

# ---------------------------------------------------------
# 10. Evaluate ALL combinations
# ---------------------------------------------------------
from itertools import combinations

# Put all your datasets in one dictionary
feature_sources = {
    "mordred": mordred_df,
    "maccs": maccs_df,
    "2d": padel2d_features,
    "3d": padel3d_features
}

results = {}

# Generate all combinations of size 2 to N
all_keys = list(feature_sources.keys())
for r in range(2, len(all_keys) + 1):
    for combo in combinations(all_keys, r):
        combo_name = " + ".join(combo)

        # Merge the feature DataFrames for this combo
        merged_df = pd.concat([feature_sources[k] for k in combo], axis=1)

        # Train + evaluate your model
        results[combo_name] = evaluate_model(merged_df, y)

# Print results
print(f"\n===== FINAL RESULTS ({MODEL_TYPE.upper()}) =====")
for k, v in results.items():
    print(k, "->", v)

```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models

# --------------------
# 1. Load data
# --------------------
labels_df = pd.read_excel("abs_Caco2_Wang.xlsx")
desc_df = pd.read_excel("abs_Caco2_Wang_mordred_desc.xlsx")

# Reset index if needed
if desc_df.index.name is not None:
    desc_df = desc_df.reset_index()

# Merge on common key
common_cols = set(labels_df.columns) & set(desc_df.columns)
if not common_cols:
    raise ValueError("No common column to merge on")
merge_key = list(common_cols)[0]

labels_df[merge_key] = labels_df[merge_key].astype(str)
desc_df[merge_key] = desc_df[merge_key].astype(str)

merged_df = pd.merge(labels_df, desc_df, on=merge_key, how="inner")

# --------------------
# 2. Features and labels
# --------------------
y = merged_df["Y"].astype(float)
X = merged_df.drop(columns=["Y", merge_key])

# Ensure numeric
X = X.apply(pd.to_numeric, errors='coerce').fillna(0)

# Remove constant features
X = X.loc[:, (X != X.iloc[0]).any()]

# Remove highly correlated features
corr = X.corr().abs()
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
drop_cols = [col for col in upper.columns if any(upper[col] > 0.95)]
X = X.drop(columns=drop_cols)

print("Final feature count:", X.shape[1])

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --------------------
# 3. Reshape for CNN
# --------------------
n_features = X_scaled.shape[1]
# make it a square (pad if needed)
side = int(np.ceil(np.sqrt(n_features)))
pad_len = side * side - n_features

X_padded = np.hstack([X_scaled, np.zeros((X_scaled.shape[0], pad_len))])
X_reshaped = X_padded.reshape(-1, side, side, 1)

# --------------------
# 4. Train-test split
# --------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_reshaped, y, test_size=0.2, random_state=42
)

# --------------------
# 5. CNN Model
# --------------------
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation="relu", input_shape=(side, side, 1)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation="relu"),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128, activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(1)  # regression output
])

model.compile(optimizer="adam", loss="mse", metrics=["mae"])
model.summary()

# --------------------
# 6. Training
# --------------------
history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=32,
    verbose=1
)

# --------------------
# 7. Evaluation
# --------------------
y_pred = model.predict(X_test).flatten()

print("R score:", r2_score(y_test, y_pred))
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE:", rmse)

# --------------------
# 8. Learning curves
# --------------------
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.legend()
plt.show()
```
